## Предобработка данных

## Проверка статистических гипотез

В последующих разделах не совсем уверена, правильно ли произвожу анализ данных

### Применение параметрического критерия

Проверка распределений на нормальность

```python
#c помощью Q-Q графика квантиль-квантиль график
fig = plt.figure(figsize=(14,5))
stats.probplot(female_wilks, dist="norm", plot=plt) #сравнивает распределения
plt.show()
fig = plt.figure(figsize=(14,5))
stats.probplot(male_wilks, dist="norm", plot=plt)
plt.show()
```

Гипотеза $ H_0 $: Данные распределены нормально. 

Так как наблюдений больше, чем 5000, применим критерий Колмогорова-Смирнова

```python
print("wilks женщин:", stats.kstest(female_wilks, 'norm'))
print("wilks мужчин:", stats.kstest(male_wilks, 'norm'))
```

Критерий Шапиро-Уилка: https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.shapiro.html

```python
print("wilks женщин:", stats.shapiro(female_wilks[:5000]))
print("wilks мужчин:", stats.shapiro(male_wilks[:5000]))
```

Вывод: распределение не является нормальным, значит нельзя применять параметрические критерии! :(

Применим критерий для доли (Z-критерий для несвязанных выборок). 

https://www.statsmodels.org/dev/generated/statsmodels.stats.proportion.proportions_ztest.html

Вычислим средний wilks (mean) по всем спортсменам. Разделим мужчин и женщин на группы: wilks < mean и wilks > mean. 

$H_0 :$ доли мужчин и женщин, у которых wilks > mean равны

$H_1 :$ доля мужчин больше (кажется, что верна альтернативная гипотеза - проверим)

```python
import random
from statsmodels.stats.proportion import proportions_ztest
#norm_vel_sample = norm_vel.rvs(size=50)
#norm_sam_sample = norm_sam.rvs(size=50)
mean_m = np.mean(male_wilks)
mean_f = np.mean(female_wilks)

male_cnt = male_wilks[male_wilks > mean_m].shape[0]
female_cnt = female_wilks[female_wilks > mean_f].shape[0]
count = np.array([female_cnt, male_cnt])
nobs = np.array([female_wilks.shape[0], male_wilks.shape[0]])
proportions_ztest(count, nobs, alternative='smaller')
```

Нулевая гипотеза отвергается. Вывод: доля мужчин с wilks выше среднего, больше, чем аналогичная доля среди женщин

### Применение непараметрического критерия

Задача - проверить гипотезу о том, что wilks у мужчин больше, чем у женщин. Применяем критерий Манна-Уитни (выборки не связаны)

Если нулевая гипотеза справедлива, то каждый из рангов с одинаковой вероятностью мог реализоваться как в выборке X1, так и в выборке X2.

```python
stats.mannwhitneyu(female_wilks, male_wilks, alternative = 'greater')
```

Нулевая гипотеза отвергается

## Предсказание временного ряда

```python
import datetime
```

Посмотрим, как менялось число женщин-участниц в соревнованиях в зависимости от времени проведения соревнований

```python
df['Date'] = pd.to_datetime(df.Date, format="%Y-%m-%d")
```

```python
df1 = df[['Date', 'Sex']]
df1 = df1[df.Sex == 'F']
df1 = pd.DataFrame(df1['Date'].value_counts())
df1.Date.plot(figsize=(15,6),title='Число женщин-участниц', fontsize=14)
```

```python
df1 = df1.sort_index()
```

```python
fig, axes = plt.subplots(1, 2, sharey=False, sharex=False)
fig.set_figwidth(14)
fig.set_figheight(4)

window = 30
axes[0].plot(df1.index, df1, label='Original')
axes[0].plot(df1.index, df1.rolling(window=window, min_periods=1).mean(), label='1-Month Rolling Mean')
axes[0].set_title("1-Month Rolling Mean")
axes[0].legend(loc='best')

window = 30*6
axes[1].plot(df1.index, df1, label='Original')
axes[1].plot(df1.index, df1.rolling(window=window, min_periods=1).mean(), label='6-Month Rolling Mean')
axes[1].set_title("6-Month Rolling Mean")
axes[1].legend(loc='best')

plt.tight_layout()
plt.show()
```

```python
from pylab import rcParams
import statsmodels.api as sm
rcParams['figure.figsize'] = 12, 7
result = sm.tsa.seasonal_decompose(df1.Date, model='additive',freq=30*6)
result.plot()
plt.show()
```

Таким образом, получилось извлечь из ряда тренд. С каждым годом число женщин-участниц растет!

Наибольший Wilks в зависимости от месяца (года)

```python
df1 = df[df.Date > datetime.datetime(2016, 1, 1)]
```

```python
df1 = df1.groupby(by='Date')['Wilks'].agg([np.max])
df1.head()
```

```python
df1.amax.plot(figsize=(15,6),title='Максимальный Wilks', fontsize=14)
```

```python
from statsmodels.tsa.stattools import adfuller
def test_stationarity(timeseries, window = 24, cutoff = 0.05):

    #Determing rolling statistics
    rolmean = timeseries.rolling(window).mean()
    rolstd = timeseries.rolling(window).std()

    #Plot rolling statistics:
    fig = plt.figure(figsize=(12, 8))
    orig = plt.plot(timeseries, color='blue',label='Original')
    mean = plt.plot(rolmean, color='red', label='Rolling Mean')
    plt.legend(loc='best')
    plt.title('Rolling Mean & Standard Deviation')
    plt.show()
    std = plt.plot(rolstd, color='black', label = 'Rolling Std')
    
    plt.show()
    
    #Perform Dickey-Fuller test:
    print('Results of Dickey-Fuller Test:')
    dftest = adfuller(timeseries.values,autolag='AIC' )
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value
    pvalue = dftest[1]
    if pvalue < cutoff:
        print('p-value = %.4f. The series is likely stationary.' % pvalue)
    else:
        print('p-value = %.4f. The series is likely non-stationary.' % pvalue)
    
    print(dfoutput)
```

```python
sample_data = df1.amax
test_stationarity(sample_data)
```

Временной ряд стационарен!

Попробуем сделать предсказание

```python
from statsmodels.tsa.api import ExponentialSmoothing
train = df1.amax.iloc[:240]
test = df1.amax.iloc[240:]
train.iloc[-110:].plot( figsize=(15,6),title='Максимальный Wilks', fontsize=14)
test.plot(figsize=(15,6),title='Максимальный Wilks', fontsize=14)
forecast = test.copy()

fit1 = ExponentialSmoothing(np.asarray(train), seasonal_periods=24, trend='add', seasonal='add').fit()
forecast = pd.Series(fit1.forecast(len(test)))
forecast.index = test.index
forecast.plot(figsize=(15,6), title='Максимальный Wilks', fontsize=14)
```

```python
rcParams['figure.figsize'] = 20, 5
df1['year'] = df1.index.year
df1['month'] = df1.index.month
df1_pivot = pd.pivot_table(df1, values = "amax", columns = "year", index = "month")
df1_pivot.plot()
```

По всей видимости, характеристика "Максимальный wilks" особо не меняется со временем. Можно проследить некоторую сезонность, максимум приходится на весну, возможно, из-за того, что весной часто проводятся соревнования.
